```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Example Hindi text
hindi_text = "यह एक उदाहरण हिंदी टेक्स्ट है। यह शब्द बादल बनाने के लिए है।"
```

```python
# Specify the font path or name here
font_path = '/home/arnav/BTP/Arnav_Medha/CSE556-NLP-Project/NER-Experiments/NotoSansDevanagari-VariableFont_wdth,wght.ttf'
```

```python
# Create a word cloud object with Hindi font support
wordcloud = WordCloud(font_path=font_path, width=800, height=400, background_color='white').generate(hindi_text)
```

```python
# Display the word cloud
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Turn off axis numbers and ticks
plt.show()
```
